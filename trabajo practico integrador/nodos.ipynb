```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Crear un grafo dirigido
G = nx.DiGraph()

# 2. Añadir 8 nodos
nodos = list(range(1, 9))
G.add_nodes_from(nodos)

# 3. Añadir relaciones dirigidas (cada nodo se conecta a los 2 siguientes en forma de anillo)
# Esto garantiza que cada nodo tenga exactamente 2 conexiones de salida (comunicación directa)
for i in nodos:
    # Primer vecino (el siguiente nodo)
    vecino1 = i + 1 if i < 8 else 1
    # Segundo vecino (saltando un nodo)
    vecino2 = i + 2 if i < 7 else (1 if i == 7 else 2)

    G.add_edge(i, vecino1)
    G.add_edge(i, vecino2)

# 4. Configurar el diseño visual en forma de círculo para apreciar la estructura
pos = nx.circular_layout(G)

# 5. Dibujar el grafo
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue')
nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif', font_weight='bold')

# Dibujar las flechas (relaciones direccionadas)
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray', connectionstyle='arc3,rad=0.1')

plt.title("Red Dirigida de 8 Nodos (Cada nodo se comunica con 2)", fontsize=14)
plt.axis('off')
plt.show()
```
